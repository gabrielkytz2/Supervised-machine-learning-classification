### PART 2) Classification Models

### Objectives

The main objective of this work is to create supervised machine learning models to classify breast cancer as benign or malign. Then, three different classification models were created in order to identify the best class of model to make cancer predictions. Classes: Support Vector Machines, Logistic Regression, and Random Forests (bagging). 

The best models of all classes was selected based on classification accuracy metrics like Accuracy, precision, recall, fscore, and AUC-ROC. All models were analyzed in both training and validation sets to identify overfitting, when it occurs. The dataset were balanced by class. Cross validation was performed to see model's stability in predictions and a grid search was performed to identify the best hiperparameters to tune the model.

Recall score (tp / (tp + fn)) was considered to select to ranking the models of each class during the Cross Validation grid search. Recall is considered the ability of the classifier to find all the positive samples, which is extremely important in the context o cancer. False negative predictions can delay patient's treatment, and time is key in cancer scenario.

The best model may be used to predict new cases of breast cancer in patient's, assisting in medical diagnosis.

### Preprocessing data

```python
#Balancing the dataset

df_neutral = df[df['diagnosis'] == 'M'] #Selecting only neutral
df_deleterious = df[df['diagnosis'] == 'B'] #Selecting only deleterious

print ('Neutral:', len(df_neutral)) # Seeing length
print ('Deleterious: ', len(df_deleterious)) # Seeing length
```

```python
df_deleterious = df_deleterious.sample(len(df_neutral)) # Downsampling deleterious data

df_final = pd.concat ([df_neutral, df_deleterious]) #Concatenating datasets again

df_final = df_final.reset_index().drop ('index', axis =1) #Reseting index

df_final # Seeing the balanced data
```

```python
# Scaling the data, needed for SVC
from sklearn.preprocessing import MinMaxScaler

X = df_final[df_final.columns[-1]] #not scaled X

y = df_final[df_final.columns[-1]] #y variable

X_scaled = df_final[df_final.columns[:-1]]
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_scaled)
X_scaled = pd.DataFrame(X_scaled, columns=df_final.columns[:-1]) #Scaled X

X_scaled #Seeing the scaled X data
```

```python
#Adjusting y column to store binnary variable 0 (benign) and 1 (malign)
import re 
y = y.map (lambda x: re.sub ('M', '1', x))
y = y.map (lambda x: re.sub ('B', '0', x))
y = y.map (lambda x: int (x))
```

```python
#Spliting the data into train and test
from sklearn.model_selection import train_test_split

#70% train
#30% test

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=0, stratify = y)
```

### A) Support vector machine classifier

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

# Parameters to grid seach
param_grid = {'C': [0.01, 0.1, 1, 10, 20]} #Setting different regularization parameter L2

# Grid search function
SVC = GridSearchCV(SVC(random_state=0,
                          kernel = 'rbf'),param_grid=param_grid, 
                          scoring='recall',n_jobs=-1, cv=4)

# Performing the grid search
SVC = SVC.fit(X_train, y_train)

#Saving model to a new variable
best_SVC = SVC.best_estimator_

# Performing the grid search
best_SVC #Seeing the best model's hiperparameters According to Recall
```

### Analyzing classification metrics for SVC model

```python
#Defining a function to print classification metrics

def classification_report (model, X_test, y_test, X_train, y_train): 
    
    from sklearn.metrics import precision_recall_fscore_support as score
    from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score, plot_roc_curve
    
    y_pred = model.predict (X_test)
    
    precision, recall, fscore, _ = score(y_test, y_pred, average='weighted')
        
    # The usual way to calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    
    # Last, the confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    ax = sns.heatmap(cm, annot=True, fmt='d')
    
    #Metrics
    metrics = []
    metrics.append(pd.Series({'precision':precision, 'recall':recall, 
                                  'fscore':fscore, 'accuracy':accuracy}))
    
    dataframe = pd.DataFrame (metrics).T
    dataframe = dataframe.rename ({0:'Metrics'}, axis = 1)
    
    #ROC AUC
    
    #Train
    plot_roc_curve(model, X_train, y_train)
    
    #Test
    plot_roc_curve(model, X_test, y_test)
    
    print ('Metrics order: \n 1) Dataframe, \n 2) Confusion matrix, \n 3) AUC ROC plot Train, \n 4) AUC ROC plot Test')
    
    return dataframe
    

classification_report (best_SVC, X_test, y_test, X_train, y_train)
```

### B) Logistic regression model

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

# Parameters to grid seach

#Setting different regularization type
#Setting different regularization parameter efficiency
param_grid = {'penalty': ['l1','l2','elasticnet'], 
              'C': [0.01, 0.1, 1, 10, 20]} 

# Grid search function
LR = GridSearchCV(LogisticRegression(random_state=0,
                          solver = 'lbfgs'),param_grid=param_grid, 
                          scoring='recall',n_jobs=-1, cv=4)

# Performing the grid search
LR = LR.fit(X_train, y_train)

#Saving model to a new variable
best_LR = LR.best_estimator_

# Performing the grid search
best_LR #Seeing the best model's hiperparameters According to Recall
```

### Analyzing classification metrics for LR model

```python
classification_report (best_LR, X_test, y_test, X_train, y_train)
```

### B) Random Forests

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Parameters to grid seach

#Setting different regularization type
#Setting different regularization parameter efficiency
param_grid = {'n_estimators': [100, 200, 300], 
              'max_depth': [5, 7, 10],
              'max_features': [10, 20, 30]} 

# Grid search function
RF = GridSearchCV(RandomForestClassifier(random_state=0),
                                         param_grid=param_grid, 
                                         scoring='recall',
                                         n_jobs=-1, cv=4)

# Performing the grid search
RF = RF.fit(X_train, y_train)

#Saving model to a new variable
best_RF = RF.best_estimator_

# Performing the grid search
best_RF #Seeing the best model's hiperparameters According to Accuracy
```

### Analyzing classification metrics for RF model

```python
classification_report (best_RF, X_test, y_test, X_train, y_train)
```

### A) Best Model

According to the classification metrics shown above, i.e. precision, recall, fscore, accuracy, auc_roc, the best model to predict whether cancer is benign or malign is Support Vector Machine model. Best values for these metrics is 1.

SVC model presented the highest precision (0.980392), then it presented the ability to do not label as positive a sample that is negative (tp / (tp + fp)). 

The model also presented the highest recall (0.979592), which is extremely important in the context of cancer (tp / (tp + fn)). A pacient with cancer needs treatment as soon as possible, and a false negative predictions are extremely harmful, since it delays treatment. 

The SVC was also the model with the highest accuracy (0.979592), so the model was able to predict true positive and true negative values more precisely than the other models (TP+TN/TP+TN+FP+FN). 

The model also presented the highest fscore metric (0.979583), which is calculated by F1 = 2 * (precision * recall) / (precision + recall). Fscore measure the precision/recall balance of the model.

At last, the model also presented the highest and maximum AUC ROC curve (1), which indicates that the model performs better in terms of true predictions and false predictions trade-off, which is again desirable in the scenario of cancer. False negative predictions are bad for patient's treatment, since it delays it. On the other hand, false positive predictions are also bad, since it distress a lot the patient unnecessarily.

### B) Main Findings

As discussed above, the SVC model presented very high classification metrics (all of them close to 1, which is the maximum possible value). An important metric in cancer diagnosis scneario is recall, i.e. the model's ability to identify all Positive values among the sample. False negative predictions are particularly harmful in cancer diagnosis, since it delays it, which can worsen the disease a lot. The SVC model shown good Recall values (0.979592), thus the model can identify approximately 98% of all positive cases, a very high proportion. Moreover, the high AUC_ROC value (1), which implicates in very low (near to 0) false positive and false negative rate, also suggest that the model is very efficient in distinguish malignant cancers from those benigns. Thus, our metrics suggests that the SVC model is a very precise and efficient model to distinguish a malignant cancers from a benign one, avoiding false negative predictions very well. The SVC model can efficiently help doctors to make correct diagnosis on breast cancer.

### C) Model's limitation and perspectives 

Despite being a very good model given the metrics analyzed, the SVC model is not perfect, since it was not abble to predict with 100% accuracy all the true positive samples (malignant breast cancer). We observed a 98% recall score, indicating that the model is able to correctly detect 98% of the malignanct cancers. But the model still presents a false negative rate of 2%, which can be improved. 

Another weakness of the model was the low number of samples included. The initial dataset was relatively short, with 569 rows. This number decreases after all data processing. Only 326 samples were included in the final model given that NA values and outliers were removed, as well as some malignant samples due to downsampling to balance the classes. 

Then a possible strategy to improve the model should be to increase the number of initial samples, or even upscaling the benign class, instead of downscaling the malignant class.

Other possible strategy should be using other classes of supervised machine learning models, such as neural network classifiers, gradient boosting classifiers or even stacking (voting) classifiers, merging different types of models. These methods are usually efficient in classifiyng binary samples, like those presented in this work.
